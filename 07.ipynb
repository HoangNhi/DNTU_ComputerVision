```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
my_path = 'C:/Users/hoang/Desktop/Image/MyRoom.jpg'
img = Image.open(my_path)
# Display(img)
plt.imshow(img)
```

```python
# Vẽ đồ thị cơ bản
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 25, 30]
# Vẽ biểu đồ có màu đỏ
plt.plot(x, y, color = "red")
# Set Title
plt.title("Ví dụ về biêu đồ")
# Tắt hệ trục tọa độ
plt.axis("off")
# Display
plt.show()
```
