```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'C:/Users/hoang/Desktop/Image/MyRoom.jpg'
img = Image.open(my_path)
plt.imshow(img)
```

```python
img_gray = img.convert("L")
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Chuyển hình ảnh thành mảng Array
img_array = np.array(img_gray)
print(img_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
# flatten: Chuyển mảng 2 chiều thành mảng 1 chiều
plt.hist(img_array.flatten(), bins = 128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel("Giá trị của Pixel")
plt.ylabel("Số lượng")
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
# Nhìn thấy đường viền
plt.contour(img_gray, origin="image")
plt.show()
```
