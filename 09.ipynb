```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'C:/Users/hoang/Desktop/Image/MyRoom.jpg'
img = Image.open(my_path)
plt.imshow(img)
```

```python
img_gray = img.convert("L")
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Chuyn hnh nh thnh mng Array
img_array = np.array(img_gray)
print(img_array)
```

```python
# To mt hnh v mi
plt.figure()
# flatten: Chuyn mng 2 chiu thnh mng 1 chiu
plt.hist(img_array.flatten(), bins = 128)
plt.title('Biu  Histogram ca nh')
plt.xlabel("Gi tr ca Pixel")
plt.ylabel("S lng")
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
# Nhn thy ng vin
plt.contour(img_gray, origin="image")
plt.show()
```
