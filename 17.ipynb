```python
from scipy import ndimage
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Đọc ảnh
image_path = 'C:/Users/hoang/Desktop/Image/architecture.jpg'
image = Image.open(image_path)

plt.imshow(image)
```

```python
# Ảnh xám
gray_image = image.convert("L")

# Chuyển ảnh xám về mảng
image_a = np.array(gray_image)

# Tính gradient theo hướng x
gradient_x = ndimage.sobel(image_a, axis = 0, mode = 'constant')

# Tính gradient theo hướng y
gradient_y = ndimage.sobel(image_a, axis = 1, mode = 'constant')

# Tính gradient theo hướng x (đạo hàm riêng theo x) bằng mặt nạ Prewitt x
gradient_x_1 = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tính gradient theo hướng y (đạo hàm riêng theo y) bằng mặt nạ Prewitt y
gradient_y_2 = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tính biên độ gradient
gradient_magnitude = np.sqrt(gradient_x_1**2 + gradient_y_2**2)

# Hiển thị ảnh
fig, axis = plt.subplots(2, 2, figsize = (10, 8))

plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap= 'gray')
plt.title("Ảnh gốc")

plt.subplot(2, 2, 2)
plt.imshow(gradient_x, cmap= 'gray')
plt.title("Bộ lọc Sobel theo hướng x")

plt.subplot(2, 2, 3)
plt.imshow(gradient_y, cmap= 'gray')
plt.title("Bộ lọc Sobel theo hướng y")

plt.subplot(2, 2, 4)
plt.imshow(gradient_magnitude, cmap= 'gray')
plt.title("Bộ lọc Prewitt")
```

```python
# Ảnh xám
gray_image = image.convert("L")

# Chuyển ảnh xám về mảng
image_a = np.array(gray_image)

# Sử dụng bộ lọc Sobel
gradient_x = ndimage.sobel(image_a, axis=0)
gradient_y = ndimage.sobel(image_a, axis=1)
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hiển thị ảnh
fig, axis = plt.subplots(2, 1, figsize = (10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap= 'gray')
plt.title("Ảnh gốc")

plt.subplot(2, 1, 2)
plt.imshow(gradient_magnitude, cmap= 'gray')
plt.title("Bộ lọc Sobel")
```
