```python
import numpy as np
from PIL import Image

def average_images(image_list):
    total_array = np.array(Image.open(image_list[0]), 'f')
    count=1
    first_image = Image.open(image_list[0])
    size = first_image.size

    for image_path in image_list[1:]:
        try:
            # Thm resize()  cho cc hnh nh c cng kch thc
            image_array = np.array(Image.open(image_path).resize(size), 'f')
            total_array += image_array
            count=count+1
        except:
            # Bt cc li nh: Cc hnh nh khng cng kch thc,...
             print("Skip ", image_path);

    # Tnh trung bnh
    average_array = total_array / count

    # Chuyn i kt qu trung bnh thnh hnh nh
    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image

# S dng v d:
my_path_1 = 'C:/Users/hoang/Desktop/Image/skywithsword.jpg'
my_path_2 = 'C:/Users/hoang/Desktop/Image/blackhole.jpg'
image_list = [my_path_1, my_path_2]
result_image = average_images(image_list)
display(result_image)


# # S dng v d:
# my_path_1 = 'C:/Users/tungi/OneDrive/Documents/src_cv/img/01/1.jpg'
# my_path_2 = 'C:/Users/tungi/OneDrive/Documents/src_cv/img/01/2.jpg'
# image_list = [my_path_1, my_path_2]
# result_image = average_images(image_list)
# display(result_image)
```
