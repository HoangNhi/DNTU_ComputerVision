```python
#Import thư viện
from PIL import Image
```

```python
# Thư mục chứa ảnh
# Trong python dùng dấu / thay cho \\
import os
print(os.getcwd())
#Thư mục chứa ảnh
dir = 'C:/Users/hoang/Desktop/Image'
```

```python
# Tạo đường dẫn đến ảnh
image_path = dir + '/MyRoom.jpg'
print(image_path)
```

```python
img01 = Image.open(image_path)
```

```python
display(img01)
```

```python
#In định dạng của ảnh
print("Định dạng của ảnh: ", img01.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", img01.size)
```

```python
# Chuyển đổi ảnh thành định dạng mới
new_img_01_path = dir + "/new_01.JPG"
img01.save(new_img_01_path)
```
